```python
#This is the beggining of the "Candle-19 Project".
#The hypothesis presented is as follows:
#    As loss of sense of smell was one of the most prevalent symptoms of 
#    COVID-19, especially in the early phases, this should be reflected in
#    the ratings for scented candles.
#        Scented candles provide a great test, as the rating scale is fairly simple:
#        1. We are not interested in positive ratings, but negative ratings
#        2. Negative rating (1 star out of 5 star) indicate that the product has
#           failed to satisfy the basic requirement of the scented candle - the scent
#        3. If hypothesis is correct, there should be an increase or a bump of 1 star
#           reviews that closely follows the breakouts of COVID-19 pandemic resurgances
#           (so called waves). These should also decrease over time, as loss of sense of
#           smell stops being a predominant symptom.

#First let's find and document the top sellers in candle section of Amazon.co.uk store.
#For sake of convinience, we will only take in consideration the popular Yankee Candle 
#scented candles.

#Question 1:
#    How many of the top 10 candles are Yankee Candle?
#Question 2:
#    What is the average rating of the Yankee Candle?
#Question 3:
#    Based on the rating, which Yankee Candle(s) would make a good candidate for
#    COVID-19 correlation the hypothesis?
#        Ratings would need to extend for at least 3 years, preferably more.
```

```python
#imports -> requests and bs4 for scraping, pandas and numpy for dataframe
from bs4 import BeautifulSoup as bs
import requests
import time
import datetime
import pandas as pd
```

```python
URL = 'https://www.amazon.co.uk/Best-Sellers-Home-Kitchen-Candles/zgbs/kitchen/3028549031/ref=zg_bs_nav_kitchen_3_3028522031'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36"}
page = requests.get(URL, headers=headers)
soup = bs(page.content, "html.parser")
```

```python
#Basic preparations -> url, headers, defining
URL = 'https://www.amazon.co.uk/Best-Sellers-Home-Kitchen-Candles/zgbs/kitchen/3028549031/ref=zg_bs_nav_kitchen_3_3028522031'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36"}
page = requests.get(URL, headers=headers)
soup = bs(page.content, "html.parser")

#Extract from page: Rank, Name, Rating, n of ratings, URL
candleslist =[]
candles = soup.find_all('div', id="gridItemRoot")
for candle in candles:
#   necessary prepwork, to be called within list
    r = candle.find('div', class_="a-icon-row").find('a', class_="a-link-normal")
    Rating_base = r.get_text(strip=True).replace(' out of 5 stars', '|') if r else None
    n_of_rating_extracted = Rating_base.partition("|")[2]
#   List
    Rankings = {
    'Rank': candle.find('span', class_="zg-bdg-text").get_text().replace('#', ''),
    'Name': candle.find('div', class_="_cDEzb_p13n-sc-css-line-clamp-3_g3dy1").get_text(strip=True),
    'Rating': float(Rating_base.partition("|")[0]),
    'n_of_rating': n_of_rating_extracted.replace(',','')
    }
    candleslist.append(Rankings.copy())
```

```python
# Loading up the data onto dataframe
df = pd.DataFrame(candleslist).set_index('Rank')
```

```python
#Fetch links as List
links = soup.find_all("a", attrs={'class':'a-link-normal'})
#Store
links_list = []
#Loop and append
for link in links:
    links_list.append(link.get('href')) 
df_links = pd.DataFrame()
df_links = df_links.append(links_list)
df_links = df_links[df_links[0].str.startswith('/product')].set_index(df.index)
df_links
```

```python
df['Links'] = df_links[0]
```
